Dowanload datasets and assign to train and testing parts

```python
(x_train, y_train), (x_test, y_test) = K.datasets.cifar10.load_data()
```

Take a peek on the images in dataset

```python
from matplotlib import pyplot as plt

def draw(x, y, row, col, n):
    p=plt.subplot(row, col, n)    
    p.set_title(y)
    plt.imshow(x)

show_size = 10
plt.figure(figsize=(20,20))

class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

for i in range(show_size):
    draw(x_train[i], class_names[y_train[i][0]], 1, show_size, i+1)
plt.show()
```
